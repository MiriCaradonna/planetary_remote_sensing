## Validation Test 1: Homogeneous Atmosphere

```python
# Defining the analytical solution
def analytical_transmission(opacity, emission_angle, H):
    return np.exp(-opacity * H / np.cos(emission_angle))
```

```python
# Define the uniform opacity profile
opacity_profile = np.full(101, 0.1) * 0.001  # 0.1 km^-1 from 0 to 100 km converted to m
altitude_grid = np.linspace(0, 100, 101) * 1000  # Convert km to meters

# Set parameters
surface_alt = 0  # Surface altitude in meters
planet_radius = 6371e3  # Earth radius in meters

# Calculating transmission vs. emission angle using the numerical method
emission_angles = np.arange(0, 90, 1)
transmissions = []
for angle in emission_angles:
    angle_rad = np.radians(angle)
    transmission = rt.surf_transmission(opacity_profile, angle_rad, 100e3, planet_radius, altitude_grid)
    transmissions.append(transmission)
```

```python
# Calculating transmission vs. emission angle using the analytical method
H = 100e3 # scale height in m
opacity = (0.1*0.001) # opacity in m

analytical_transmissions = []

for angle in emission_angles:
    angle_rad = np.radians(angle)
    transmission_analytical = analytical_transmission(opacity, angle_rad, H)
    analytical_transmissions.append(transmission_analytical)
```

```python
# Plotting transmission vs. emission angle for both the numerical and analytical solutions
plt.figure(figsize=(10, 6))
plt.plot(emission_angles, analytical_transmissions, color = 'red', label = 'Analytical Transmission')
plt.plot(emission_angles, transmissions, color = 'black', label='Numerical Transmission', linestyle = '--')
plt.xlabel('Emission Angle (degrees)')
plt.ylabel('Transmission')
plt.title('Transmission vs. Emission Angle')
plt.legend()
```

#### Discussion: Numerical vs. Analytical

The numerical and the and the analytical solutions match up pretty well! They deviate very slightly around 30-40 degrees, but that difference is pretty much negligible.

## Validation Test #2: Mars Water Ice Cloud

```python
#Test Parameters
cloud_alt = 15e3 #Altitude of the water ice cloud (km to m)
surf_temp = 250 #Surface temperature in Kelvin
a = 0.25 #Albedo
tau = 0.3 #optical depth
scattering_a = 0.9 #single scattering albedo
g = 0.7 #asymmetry parameter
thermal_wavelength = 12e-6 #wavelength in meters
cloud_temp = 220 #Temperature of the cloud in Kelvin
mars_radius = 3390e3 #Radius of Mars in meters

#below are values I have chosen for the problem or have gotten a value from reputable sources
solar_flux = 586.2 #W/m2; estimated incident solar flux on Mars from the NASA Mars fact sheet: 
                                        #https://nssdc.gsfc.nasa.gov/planetary/factsheet/marsfact.html

sza = 0 #solar zenith angle in radians for an overhead sun
emission_angles = np.arange(0, 90, 1)
azimuth_angle = 0 #azimuth angle in radians
```

```python
#calculating visible cloud brightness
visible_cloud_brightness = []
for angle in emission_angles:
    angle_rad = np.radians(angle)
    vis_brightness = rt.cloud_visible_brightness(a, solar_flux, sza, angle_rad, 
                                                     azimuth_angle,tau,scattering_a,g)
    visible_cloud_brightness.append(vis_brightness)
```

```python
#Calculating thermal cloud brightness
wavelengths = np.linspace(0.01e-9, 1000e-9, len(emission_angles))

thermal_cloud_brightness = []

for angle in emission_angles:
    angle_rad = np.radians(angle)
    spectral_brightness = []

    for wavelength in wavelengths:
        thermal_brightness = rt.cloud_thermal_brightness(surf_temp, thermal_wavelength, tau, scattering_a, 
                                                             angle_rad,cloud_temp,cloud_alt,mars_radius)
        spectral_brightness.append(thermal_brightness) 
    
        
    spectral_brightness = np.array(spectral_brightness)
        
    #Numerical integration using np.trapz (converting from W/m2/Sr/m to W/m2/Sr)
    total_brightness = np.trapz(spectral_brightness, wavelengths)
        
    thermal_cloud_brightness.append(total_brightness)
```

```python
fig,(ax1,ax2) = plt.subplots(ncols=2, figsize=(10, 6))
ax1.plot(emission_angles, thermal_cloud_brightness, color = 'red', label = 'Thermal Cloud Brightness')
ax2.plot(emission_angles, visible_cloud_brightness, color = 'black', label='Visible Cloud Brightness')
ax1.set_xlabel('Emission Angle (degrees)')
ax1.set_ylabel('Cloud Brightness (W/m2/Sr)')
ax1.set_title('Thermal Cloud Brightness')
ax1.legend(loc = 'upper center')
ax2.set_xlabel('Emission Angle (degrees)')
ax2.set_ylabel('Cloud Brightness (W/m2/Sr)')
ax2.set_title('Visible Cloud Brightness')
ax2.legend(loc = 'upper center')
```

#### Show that total energy is conserved

```python
#calculating transmission, absorption, and reflection
transmission = np.exp(-tau)
absorption = (1 - scattering_a) * (1 - transmission)
reflection = scattering_a * (1 - transmission)

#summing transmission, absorption, and reflection to test energy conservation
print("transmission + absorption + reflection = ", (transmission + absorption + reflection))
print("Energy is conserved!")
```
