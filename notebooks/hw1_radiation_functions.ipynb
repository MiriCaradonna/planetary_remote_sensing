### 1. Basic Validation and Bandpass Effects

```python
#Creating an instance of the Radiation class
radiation = Radiation()
```

#### Plotting Spectral Radiance for The Sun, Earth, and Europa

```python
#using planck_function to plot the blackbody spectrum of the Sun

wavelength_Sun = np.linspace(10**-7, 0.8*10**-5, 1000) #creating an array of wavelengths in meters 

temperature_Sun = 5778 #temperatue of the Sun in Kelvin
planck_func_Sun = radiation.planck_function(wavelength_Sun, temperature_Sun) #calling planck_function

plt.plot(wavelength_Sun*1e6, planck_func_Sun, color = 'Red')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Wavelength ($\mu$m)')
plt.ylabel('Spectral Radiance W m-2 Sr-1 $\mu$m-1')
plt.title('Spectral Radiance of the Sun as a Function of Wavelength (log-log)')
```

```python
#using planck_function to plot the blackbody spectrum of Earth

wavelength_Earth = np.linspace(2*10**-6, 10**-4, 1000) #creating an array of wavelengths in meters 

temperature_Earth = 300 #temperatue of the Earth in Kelvin
planck_func_Earth = radiation.planck_function(wavelength_Earth, temperature_Earth) #calling the planck function

plt.plot(wavelength_Earth*1e6, planck_func_Earth, color = 'Green')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Wavelength ($\mu$m)')
plt.ylabel('Spectral Radiance W m-2 Sr-1 $\mu$m-1')
plt.title('Spectral Radiance of Earth as a Function of Wavelength (log-log)')
```

```python
#using planck_function to plot the blackbody spectrum of Europa

wavelength_Europa = np.linspace(0.5*10**-5, 5*10**-4, 1000) #creating an array of wavelengths in meters 

temperature_Europa = 100 #temperatue of the Sun in Kelvin
planck_func_Europa = radiation.planck_function(wavelength_Europa, temperature_Europa) #calling planck_function

plt.plot(wavelength_Europa*1e6, planck_func_Europa, color = 'Blue')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Wavelength ($\mu$m)')
plt.ylabel('Spectral Radiance (W m-2 Sr-1 $\mu$m-1)')
plt.title('Spectral Radiance of Europa as a Function of Wavelength (log-log)')
```

#### Testing the inverse planck function

```python
temperature_inv_Sun = radiation.planck_inv(wavelength_Sun, planck_func_Sun)
temperature_inv_Sun[0:250]
```

```python
temperature_inv_Earth = radiation.planck_inv(wavelength_Earth, planck_func_Earth)
temperature_inv_Earth[0:250]
```

```python
temperature_inv_Europa = radiation.planck_inv(wavelength_Europa, planck_func_Europa)
temperature_inv_Europa[0:250]
```

#### Band-integrated Radiance as a function of wavelength and bandwidth

```python
#attempting to use realistic bandwidths
band_center_Sun = 15e-9 #in meters
band_width_Sun = np.linspace(10e-9, 29e-9, 50) #in meters

radiance_Sun = []
for bandwidth in band_width_Sun:
    rad = radiation.radiance(temperature_Sun, band_center_Sun, bandwidth)
    radiance_Sun = np.append(radiance_Sun, rad)
```

```python
plt.plot(band_width_Sun, radiance_Sun)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('bandwidth (m)')
plt.ylabel('Radiance (W m-2 Sr-1)')
plt.title('Radiance of the Sun as a function of Bandwidth (log-log)')
```

```python
band_center_Earth = 15e-7 #in meters
band_width_Earth = np.linspace(1e-7, 29e-7, 50) #in meters

radiance_Earth = []
for bandwidth in band_width_Earth:
    rad = radiation.radiance(temperature_Earth, band_center_Earth, bandwidth)
    radiance_Earth = np.append(radiance_Earth, rad)
    
```

```python
plt.plot(band_width_Earth, radiance_Earth)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('bandwidth (m)')
plt.ylabel('Radiance (W m-2 Sr-1)')
plt.title('Radiance of Earth as a function of Bandwidth (log-log)')
```

```python
band_center_Europa = 15e-7 #in meters
band_width_Europa = np.linspace(1e-7, 29e-7, 50) #in meters

radiance_Europa = []
for bandwidth in band_width_Europa:
    rad = radiation.radiance(temperature_Europa, band_center_Europa, bandwidth)
    radiance_Europa = np.append(radiance_Europa, rad)
```

```python
plt.plot(band_width_Europa, radiance_Europa)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('bandwidth (m)')
plt.ylabel('Radiance (W m-2 Sr-1)')
plt.title('Radiance of Europa as a function of Bandwidth (log-log)')
```
